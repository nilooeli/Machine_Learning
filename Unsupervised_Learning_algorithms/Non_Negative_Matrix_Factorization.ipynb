```python
# Step 1: Import Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import NMF

# Step 2: Load the Digits Dataset
digits = load_digits()
X = digits.data
print(X.shape)

# Step 3: Applly NMF
nmf = NMF(n_components=16, random_state=42, max_iter=500)
W = nmf.fit_transform(X)
H = nmf.components_

# Step 4: Visualize the components (Parts of Digits)
fig, axes = plt.subplots(2, 8, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(H[i].reshape(8, 8), cmap = 'gray')
    ax.set_title(f'component {i}')
    ax.axis('off')
plt.suptitle('Parts of Digits Learned by NMF')
plt.tight_layout()
plt.show()
```
